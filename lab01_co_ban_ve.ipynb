### 1.1 NUMPY 

```python
import numpy as np

#Tạo một mảng 1 chiều
mang_1D = np.array([1, 2, 3, 4, 5, 6, 7, 8])
print(mang_1D)
# Tạo một mảng 2 chiều
mang_2D = np.array(([1,3,5,7],[2,4,6,8]))
#Tạ mảng nghẫu nhiên
mang_nD = np.random.rand(4,4)
print(mang_nD)
```

## các thao tác cơ bản trong numpynumpy

```python
#Tạo một mảng
mang_1D = np.array([1,2,3,4,5,6,7,8])
print(mang_1D)
#Cộng và nhân mảng với một số
print(mang_1D + 5)
print(mang_1D * 5)
#Tính tổng, trung bình, lớn nhất, nhỏ nhất
print(mang_1D.sum())
print(mang_1D.mean())
print(mang_1D.max())
print(mang_1D.min())
```

## PANDAS
### 2.1 tạo dataFrame 

```python
import pandas as pd 

#Tạo dataFrame và dictionary
data = {
    "Tên" : ["Anh", "Yêu", "Em"],
    "Tuổi" : [2006, 2007, 2009],
    "Điểm" : [7, 8, 6]
}

df = pd.DataFrame(data)
print(df)
```

### 2.2 Thao tác cơ bản

```python
print(df["Tuổi"])
#Lấy hạng thứ 2
print(df.iloc[1])

#Lấy ra người có tuổi lớn hơn 18
print(df[df["Tuổi"]>18])

#Tính điểm trung bình của người đó
```

## MATPLOTLIB
### Biểu đồ đường

```python
import matplotlib.pyplot as plt

x = [5, 8, 7, 4, 32, 99]
y = [20, 87, 64, 55, 30, 100]

plt.plot(x,y, label="Biểu đồ đường", color="pink", marker="o")
plt.title("Biểu đồ đường")
plt.xlabel("Trực X")
plt.ylabel("Trực Y")
plt.legend()
plt.show()
```

### Biểu đồ cột

```python
Ten = ["Hải", "Ami", "Fouk", "Nhân", "Koukkai"]
Diem = [7, 8.5, 9, 9.3, 8]

plt.bar(Ten, Diem, color = "Yellow")
plt.title("Biểu đồ cột")
plt.xlabel("Ten")
plt.ylabel("Diem")
plt.show()
```

### Biểu đồ phân tán

```python
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.rand(100)*100

plt.scatter(x, y, s=sizes, alpha = 0.5, c= "green")
plt.title("Biểu đồphaan tán")
plt.xlabel("Trục x")
plt.xlabel("Trục x")
plt.ylabel("Trục y")
plt.show()
```
